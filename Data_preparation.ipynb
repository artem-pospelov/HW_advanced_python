```python
import pandas as pd
```

- AGREEMENT_RK — уникальный идентификатор объекта в выборке;
- TARGET — целевая переменная: отклик на маркетинговую кампанию (1 — отклик был зарегистрирован, 0 — отклика не было);
- AGE — возраст клиента;
- SOCSTATUS_WORK_FL — социальный статус клиента относительно работы (1 — работает, 0 — не работает);
- SOCSTATUS_PENS_FL — социальный статус клиента относительно пенсии (1 — пенсионер, 0 — не пенсионер);
- GENDER — пол клиента (1 — мужчина, 0 — женщина);
- CHILD_TOTAL — количество детей клиента;
- DEPENDANTS — количество иждивенцев клиента;
- PERSONAL_INCOME — личный доход клиента (в рублях);
- LOAN_NUM_TOTAL — количество ссуд клиента;
- LOAN_NUM_CLOSED — количество погашенных ссуд клиента.

```python
df_loan_hist = pd.merge(df_loan, df_close_loan, on = 'ID_LOAN', how = 'inner')
```

```python
df_loan_hist
```

```python
df_l_count = df_loan_hist.groupby(by='ID_CLIENT').CLOSED_FL.count().reset_index()
df_l_count = df_l_count.rename(columns={"CLOSED_FL":"LOAN_NUM_TOTAL"})
df_l_count
```

```python
df_l_sum = df_loan_hist.groupby(by='ID_CLIENT').CLOSED_FL.sum().reset_index()
df_l_sum = df_l_sum.rename(columns={"CLOSED_FL":"LOAN_NUM_CLOSED"})
df_l_sum
```

```python
df = pd.merge(df, df_l_count, on = 'ID_CLIENT', how = 'left')
df = pd.merge(df, df_l_sum, on = 'ID_CLIENT', how = 'left')
```

```python
df.columns
```

```python
col = ['AGREEMENT_RK', 'TARGET', 'AGE', 'SOCSTATUS_WORK_FL', 'SOCSTATUS_PENS_FL',
       'GENDER', 'CHILD_TOTAL', 'DEPENDANTS', 'PERSONAL_INCOME','LOAN_NUM_TOTAL','LOAN_NUM_CLOSED'] #интересущие колонки
```

```python
df_agg = df[col]
```
